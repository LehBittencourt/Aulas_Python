### **Manipulando Arquivos com NumPy**

```python
import numpy
import os
filename = os.path.join('dataset.csv') #arquivo
```

```python
# No windows use !more dataset.csv. Mac ou Linux use !head dataset.csv
!more dataset.csv
```

```python
# Carregando um dataset para dentro de um array
# skiprows: pula a linha índice 0
# usecols: pega apenas as colunas desejadas
arr13 = numpy.loadtxt(filename, delimiter = ',', usecols = (0,1,2,3), skiprows = 1)
```

```python
print(arr13)
```

```python
# Carregando apenas duas variavéis (colunas) do arquivo
# unpack: ele descompacta(vai criar 2 arrays)
var1, var2 = numpy.loadtxt(filename, delimiter = ',', usecols = (0,1), skiprows = 1, unpack = True)
```

```python
# Gerando um plot a partir de um arquivo usando o NumPy
# biblioteca que permite criar gráficos
import matplotlib.pyplot as plt
plt.show(plt.plot(var1,var2, 'o', markersize = 4, color = 'red'))
```
