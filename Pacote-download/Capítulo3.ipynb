```python
# Criando duas funções

# Função 1: retorna uma temperatura em Fahrenheit
def fahrenheit(T):
    return ((float(9)/5)*T + 32)

# Função 1: retorna uma temperatura em Celsius
def celsius(T):
    return ((float(5)/9)*T - 32)
```

```python
# Usando um loop for para imprimir o resultado da função map()
for temp in map(fahrenheit, temperaturas):
    print(temp)
```

> Try, Except, Finally

```python
# Utilizando try e except
try:
    8 + 's'
except TypeError:
    print('Operação não permitida!')
```

```python
# Utilizando try, except e else - sem erro
try:
    f = open('/content/testandoErro.txt', 'w')
    f.write('Gravando erro')
except IOError:
    print('Erro: arquivo não encontrado ou não pode ser salvo!')
else:
    print('Conteúdo gravado com sucesso!')
    f.close()
```

```python
# Utilizando try, except e else - com erro (está sem o txt)
try:
    f = open('/content/testandoErro', 'r')
    f.write('Gravando erro')
except IOError:
    print('Erro: arquivo não encontrado ou não pode ser salvo!')
else:
    print('Conteúdo gravado com sucesso!')
    f.close()
```

```python
# Utilizando try, except, else e finally - sem erro
try:
    f = open('/content/testandoErro.txt', 'w')
    f.write('Gravando erro')
except IOError:
    print('Erro: arquivo não encontrado ou não pode ser salvo!')
else:
    print('Conteúdo gravado com sucesso!')
    f.close()
finally:
    print('Comandos no bloco finally são sempre executados!')
```



> Cada possibilidade de erro deve ser tratada no seu programa



```python
def askint():
    try:
        val = int(input('Digite um número:'))
    except:
        print('Você não digitou um número!')
        val = int(input('tente novamente.\nDigite um número:'))
    finally:
        print('Obrigado!')
    print(val)
```

```python
# Atenção: na primeira vez não vai quebrar, mas a partir da segunda vez que o usuário digitar errado o programa vai quebrar, pois você só tratou o erro uma vez
askint()
```

```python
def askint():
    while True:
      try:
          val = int(input('Digite um número:'))
      except:
          print('Você não digitou um número!')
          continue
      else:
          print('Obrigado por digitar um número!')
          break
      finally:
          print('Fim da execução!')
      print(val)
```

```python
# Usando o loop só vai parar o programa quando o usúario digitar corretamente
askint()
```

### **Expressões Regulares**

```python
import re
```

```python
texto = 'Meu e-mail é exemplo@gmail.com e você pode me contatar em outro_email@yahoo.com.'
```

```python
# Expressão regular para contatar quantas vezes aparece o caracter arroba no texto
# findall('caracter', x): encontra um caracter no texto
resultado = len(re.findall('@', texto))
```

```python
print("O caracter '@' apareceu", resultado, 'vezes no texto.' )
```

```python
# Expressão regular para extrair a palavra que aparece após a palavra 'você' em um texto
resultado = re.findall(r'você (\w+)', texto)
```

```python
print("A palavra após 'você' é:", resultado[0])
```

Nota: O r antes da string que representa a expressão regular em Python é usado para indicar que a string é uma string literal raw. Isso significa que as barras invertidas() não são interpretadas como caracter de escape, mas são incluídas na expressão regular como parte do padrão.

```python
# Expressão regular para extrair endereços de e-mail de uma string
emails = re.findall(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b', texto)
```

```python
print(emails)
```
